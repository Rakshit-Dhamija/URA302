```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize
from scipy.interpolate import interp1d
import scipy.optimize as opt
from scipy.optimize import fsolve
```

```python


# Q1: Student Data

names = ["Arin", "Aditya", "Chirag", "Gurleen", "Kunal"]
m_marks = [85, 78, 92, 88, 79]
p_marks = [75, 82, 74, 90, 83]
c_marks = [90, 85, 89, 78, 87]
e_marks = [88, 90, 78, 92, 85]

# Calculate total marks, average marks, subject-wise performance
total_marks = np.array(m_marks) + np.array(p_marks) + np.array(c_marks) + np.array(e_marks)
average_marks = total_marks / 4

# Identify top and bottom performers
top_performer_index = np.argmax(total_marks)
bottom_performer_index = np.argmin(total_marks)

# Determine passing percentage
passing_count = len(total_marks[total_marks >= 200])
passing_percentage = (passing_count / len(total_marks)) * 100

# Generate plots
plt.bar(names, m_marks, label="Math")
plt.bar(names, p_marks, label="Physics", bottom=m_marks)
plt.bar(names, c_marks, label="Chemistry", bottom=np.array(m_marks) + np.array(p_marks))
plt.bar(names, e_marks, label="English", bottom=np.array(m_marks) + np.array(p_marks) + np.array(c_marks))
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.title("Subject-wise Performance")
plt.legend()
plt.show()

print(f"Top Performer: {names[top_performer_index]}")
print(f"Bottom Performer: {names[bottom_performer_index]}")
print(f"Passing Percentage: {passing_percentage:.2f}%")

```

```python
# Q2: Velocity Data

time = np.array([0, 1, 2, 3, 4, 5])
velocity = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

# Define the quadratic function
def quadratic_function(t, a, b, c):
    return a * t**2 + b * t + c

# Fit the function to the data
params, _ = scipy.optimize.curve_fit(quadratic_function, time, velocity)

# Plot the original data and the fitted curve
plt.plot(time, velocity, label="Original Data")
plt.plot(time, quadratic_function(time, *params), label="Fitted Curve")
plt.xlabel("Time (seconds)")
plt.ylabel("Velocity (m/s)")
plt.legend()
plt.show()
```

```python

# Q3: Population Data

year = np.array([2000, 2005, 2010, 2015, 2020])
population = np.array([50, 55, 70, 80, 90])
plt.scatter(year,population,label="Data")

# Perform linear interpolation
f = interp1d(year, population)
year1=np.arange(2000,2021)
print(year1)
data=np.array(f(year1))
plt.plot(year1,data,label="function")
# Estimate population for the year 2008
population_2008 = f(2008)

# Plot the original data and the interpolated value
plt.scatter(2008, population_2008, color="red", label="Interpolated Value")
plt.xlabel("Year")
plt.ylabel("Population (in thousands)")
plt.legend()
plt.show()

print(f"Estimated Population in 2008: {population_2008:.2f} thousand")
```

```python
#Q4
# Define the polynomial function
def p(x):
    return 3*x**3 - 5*x**2 + 2*x - 8

# Find the roots using SciPy's fsolve
roots = fsolve(p, [-3, 0, 3])  # Initial guesses for the roots
print("Roots:", roots)

# Define the range for x
x = np.linspace(-3, 3, 400)
y = p(x)

# Plot the polynomial function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='p(x) = 3x^3 - 5x^2 + 2x - 8', color='b')

# Mark the roots on the plot
plt.scatter(roots, p(roots), color='r', zorder=5)
for root in roots:
    plt.annotate(f'Root: {root:.2f}', xy=(root, p(root)), xytext=(root, p(root)+10),
                 arrowprops=dict(facecolor='black', shrink=0.05))

# Labeling the plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Polynomial Function p(x) = 3x^3 - 5x^2 + 2x - 8')
plt.legend()
plt.show()
```

```python
#Q5
import random
import string
from datetime import datetime

# Function to generate random text
def generate_random_text(size_in_mb):
    size_in_bytes = size_in_mb * 1024 * 1024
    chars = string.ascii_letters + string.digits + string.punctuation + " "
    return ''.join(random.choices(chars, k=size_in_bytes))

# Function to write text to a file
def write_to_file(filename, text):
    with open(filename, 'w') as file:
        file.write(text)

# Function to convert text in a file to upper case
def convert_to_upper_case(filename):
    start_time = datetime.now()
    with open(filename, 'r') as file:
        data = file.read()
    data_upper = data.upper()
    with open(filename, 'w') as file:
        file.write(data_upper)
    end_time = datetime.now()
    return (end_time - start_time).total_seconds()

# File sizes in MB
file_sizes = [20, 40, 60, 80, 100]
conversion_times = []

# Generate files and convert them to upper case
for size in file_sizes:
    filename = f"random_text_{size}MB.txt"
    print(f"Generating {size}MB file...")
    random_text = generate_random_text(size)
    write_to_file(filename, random_text)
    print(f"Converting {size}MB file to upper case...")
    time_taken = convert_to_upper_case(filename)
    conversion_times.append(time_taken)
    print(f"Time taken for {size}MB file: {time_taken} seconds")

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(file_sizes, conversion_times, marker='o', linestyle='-', color='b')
plt.xlabel('File Size (MB)')
plt.ylabel('Time Taken (seconds)')
plt.title('Time Taken to Convert Text Files to Upper Case')
plt.grid(True)
plt.show()
```

```python
#Q6
from scipy.optimize import minimize

# Define the function
def f(x):
    return x**4 - 3*x**3 + 2

# Initial guess for the local minima
initial_guesses = [-1, 0, 2]

# Find the local minima using SciPy's minimize function
local_minima = []
for guess in initial_guesses:
    result = minimize(f, guess)
    if result.success:
        local_minima.append(result.x[0])

print("Local minima:", local_minima)

# Define the range for x
x = np.linspace(-2, 3, 50)
y = f(x)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^4 - 3x^3 + 2', color='b')

# Mark the local minima on the plot
plt.scatter(local_minima, f(np.array(local_minima)), color='r', zorder=5)
for xmin in local_minima:
    plt.annotate(f'Min: {xmin:.2f}', xy=(xmin, f(xmin)), xytext=(xmin, f(xmin)+20),
                 arrowprops=dict(facecolor='black', shrink=0.05))

# Labeling the plot
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function f(x) = x^4 - 3x^3 + 2 and its Local Minima')
plt.legend()
plt.show()
```
